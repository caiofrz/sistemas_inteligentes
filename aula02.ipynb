```python
import pandas as pd
```

```python
dicionario_funcionarios = {
'Nome': ['João','Carlos', 'Júlia','Clara', 'Mateus'],
'Salário': [1200., 1700., 1700.,2000.,1750.],
'QuantidadeFilhos': [0, 2, 1, 0, 3]
}
df_funcionarios = pd.DataFrame(data=dicionario_funcionarios)
df_funcionarios
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
```

```python
ax= df_funcionarios.plot.bar(
x='Nome',
y='Salário',
figsize=(10,5),
color='green',
rot=360
)
ax.set_title('Número de Dependentes')
ax.set_xlabel('Nome do Funcionário', labelpad=10)
ax.set_ylabel('Quantidade de Filhos', labelpad=10)
plt.savefig('numero_filhos.png')
```

```python
dicionario_produtos = {
'nome': ['Tenis','Pendrive', 'Placa Mãe','Camisa', 'Chinelo', 'Livro', 'Goiabada', 'Pão'],
'valor': [200., 52., 750.,150.,75., 250., 15., 7.5],
'marca': ['Nike', 'Mutilaser', 'Intel', 'Polo', 'Havaianas', 'Pearson', 'QDelicia', 'Seve'],
'categoria':['A', 'I', 'I', 'A', 'A', 'B', 'C', 'C']
}
df_produtos= pd.DataFrame(data=dicionario_produtos)
df_produtos
```

```python
df_categoria = df_produtos.groupby(['categoria'])['categoria'].count()
#df_categoria = df_produtos.groupby(['marca'])['marca'].count()
df_categoria
```

```python
ax = df_categoria.plot.pie(
title='Categoria de Produtos', #define o título do gráfico
ylabel='', #esconde o rótulo do eixo y
figsize=(10,5),
legend=True, #exibe ou oculta a legenda
autopct='%1.0f%%', #Exibir a porcentagem dos valores
)
ax.plot()
plt.savefig('categoria_produtos.png')
```
