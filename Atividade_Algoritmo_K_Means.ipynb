```python
#importando bibliotecas e conjuntos de dados necess√°rios
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns

from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

```python
# criando um array de classes que ser√£o utilizadas com base no df, excluindo a coluna 'species'
X = np.array(df.drop('species', axis=1))
X
```

```python
import sklearn
from sklearn.cluster import KMeans
```

```python
# inicializando o KMeans e definindo 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
```

```python
# treinando modelo
kmeans.fit(X)
```

```python
# mostrando a divis√£o de grupos feita pelo kmeans
kmeans.labels_
```

```python
# mostarndo a posi√ß√£o dos clusters centers
kmeans.cluster_centers_
```

```python
# adicionando mais uma coluna com os dados do agrupamento feito pelo kmeans
df['K-classes'] = kmeans.labels_
df
```

```python
# plotando o gr√°fico com base na divis√£o original de especies
sns.pairplot(df, hue='species')
```

```python
# plotando o gr√°fico com base na divis√£o de classes feitas pelo KMeans
sns.pairplot(df, hue='K-classes', palette='viridis')
```

```python
# funcao paar verificar a porcentagem de acerto do algoritmo
def accuracy(coluna_species, coluna_Kclasses):
  total_linhas = len(df)
  relacao_corretas = 0

  for i in range(total_linhas):
    if coluna_species[i] == 'Iris-setosa' and coluna_Kclasses[i] == 1:
      relacao_corretas += 1
    elif coluna_species[i] == 'Iris-versicolor' and coluna_Kclasses[i] == 0:
      relacao_corretas += 1
    elif coluna_species[i] == 'Iris-virginica' and coluna_Kclasses[i] == 2:
      relacao_corretas += 1

  return (relacao_corretas / total_linhas) * 100
```

```python
# mostrando a porcentagem de acerto
accuracy = accuracy(df['species'], df['K-classes'])
print(f"Porcentagem de acerto do algoritmo: {accuracy:.2f}%")
```

```python
# verificando e mostrando os erros
inconsistencias = df[(df['species'] != 'Iris-setosa') & (df['K-classes'] == 1)]
inconsistencias = inconsistencias.append(df[(df['species'] != 'Iris-versicolor') & (df['K-classes'] == 0)])
inconsistencias = inconsistencias.append(df[(df['species'] != 'Iris-virginica') & (df['K-classes'] == 2)])

print(f"Quantidade bruta de inconsistencias: {inconsistencias['species'].count()}")
print(inconsistencias.to_string())
```
