```python
#Colunas que queremos extrair do dataset
#Está é a relação de colunas que serão do dataset que serão liddas, caso necessite de mais al
columns = ['NU_INSCRICAO', 'NU_ANO', 'CO_MUNICIPIO_RESIDENCIA',
'NO_MUNICIPIO_RESIDENCIA', 'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA',
'NU_IDADE', 'TP_SEXO', 'TP_ESTADO_CIVIL', 'TP_COR_RACA',
'TP_NACIONALIDADE', 'CO_MUNICIPIO_NASCIMENTO',
'NO_MUNICIPIO_NASCIMENTO', 'CO_UF_NASCIMENTO', 'SG_UF_NASCIMENTO','CO_ESCOLA',
'TP_ST_CONCLUSAO', 'TP_ANO_CONCLUIU', 'TP_ESCOLA', 'TP_ENSINO','SG_UF_ESC',
'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
'NU_NOTA_MT', 'TP_STATUS_REDACAO', 'NU_NOTA_COMP1',
'NU_NOTA_COMP2', 'NU_NOTA_COMP3', 'NU_NOTA_COMP4', 'NU_NOTA_COMP5',
'NU_NOTA_REDACAO', 'Q001', 'Q002', 'Q003', 'Q004', 'Q005', 'Q006',
'Q007', 'Q008', 'Q009', 'Q010', 'Q011', 'Q012', 'Q013', 'Q014',
'Q015', 'Q016', 'Q017', 'Q018', 'Q019', 'Q020', 'Q021', 'Q022',
'Q023', 'Q024', 'Q025']
```

```python
#Descompactando dataset de microdados do enem
with zipfile.ZipFile('microdados_enem_2019.zip') as z:
  with z.open('DADOS/MICRODADOS_ENEM_2019.csv') as f:
    df = pd.read_csv(f,sep=';',encoding='ISO-8859-1', usecols=columns)
```

```python
df.head(15)
```

```python
df_tipo_escola = df.groupby(['TP_ESCOLA'])['TP_ESCOLA'].count()
df_tipo_escola
```

```python
rotulos_tipo_escola = {
1:'Não respondeu',
2:'Pública',
3:'Privada',
4:'Exterior'
}
rotulos = [rotulos_tipo_escola[tipo] for tipo in df_tipo_escola.index] #Pega o rotulo de cada
ax = df_tipo_escola.plot.pie(
labels=rotulos, #define a legenda para os dados
title='Tipo de Escola', #define o título do gráfico
ylabel='', #esconde o rótulo do eixo y
legend=True, #exibe ou oculta a legenda
autopct='%1.0f%%', #Exibir a porcentagem dos valores
)
ax.plot()
```

```python
ax =df_tipo_escola.plot.bar(
title='Tipo de Escola', #define o título do gráfico
legend=True, #exibe ou oculta a legenda
xlabel='Tipo de escola', # Define o rótulo do eixo x
ylabel='Quantidade de alunos', #define o rótulo do eixo y
rot=0 #rotação dos rótulo do eixo x
)
ax.get_legend().remove() # Esconde as legendas
ax.set_xticklabels(rotulos) #define os rótulos de cada barra
ax.plot()
```

```python
df_tipo_escola_porcentagem = pd.DataFrame(df_tipo_escola) #transforma a série em dataframe
total = df_tipo_escola_porcentagem['TP_ESCOLA'].sum() #calcula o total de inscritos
df_tipo_escola_porcentagem['porcentagem'] = df_tipo_escola_porcentagem['TP_ESCOLA']*100/total
df_tipo_escola_porcentagem
```

```python
ax =df_tipo_escola_porcentagem.plot.bar(
title='Tipo de Escola', #define o título do gráfico
legend=True, #exibe ou oculta a legenda
xlabel='Tipo de escola', # Define o rótulo do eixo x
ylabel='Quantidade de alunos', #define o rótulo do eixo y
rot=0, #rotação dos rótulo do eixo x
y='porcentagem' #define a coluna do datafrma e que vai colocada no eixo y
)
ax.get_legend().remove() # Esconde as legendas
ax.set_xticklabels(rotulos) #define os rótulos de cada barra
#Adiciona os valores no topo de de cada barra
#lista de nomes de cores: https://matplotlib.org/stable/gallery/color/named_colors.html
for i, v in enumerate(df_tipo_escola_porcentagem['porcentagem']):
  ax.text(i-0.1, v+1, '{:.2f}'.format(v), color='darkblue')
  ax.plot()
```

```python
df_tipo_sexo = df.groupby(['TP_SEXO'])['TP_SEXO'].count()
df_tipo_sexo
```

```python
rotulos_sexo = {
'M':'Masculino',
'F':'Feminino'
}
rotulos = [rotulos_sexo[tipo] for tipo in df_tipo_sexo.index] #Pega o rotulo de cada categori
ax = df_tipo_sexo.plot.pie(
labels=rotulos, #define a legenda para os dados
title='Sexo', #define o título do gráfico
ylabel='', #esconde o rótulo do eixo y
legend=True, #exibe ou oculta a legenda
autopct='%1.0f%%', #Exibir a porcentagem dos valores
)
ax.plot()
```

```python
#Dessa vez vamos usar a função pivot table para efetuar o agrupamento e efetuar a contagem
df_tipo_sexo_por_uf = df.pivot_table(
  columns=['TP_SEXO'], #Os valores desse atributo se tornaram as colunas do nosso dataframe
  index=['SG_UF_ESC'], #Os valores dessa coluna se tornaram nosso índice
  aggfunc={
  'TP_SEXO': 'count' #Os valores dessa coluna serão agregados de acordo a função de agr
  }
)
#dess forma, teremos a quantidade de inscritos por estado
df_tipo_sexo_por_uf = df_tipo_sexo_por_uf.reset_index()
df_tipo_sexo_por_uf = df_tipo_sexo_por_uf.set_index('SG_UF_ESC')
df_tipo_sexo_por_uf
```

```python
ax = df_tipo_sexo_por_uf.plot.bar(
title='Sexo dos inscritos por UF',
xlabel='', #Remove o label do eixo X
ylabel='Quantidade', #Definie o label do eixo Y
figsize=(15,5),
)
ax.plot()
```
